```python
import rasterio
import os.path
import xarray
import matplotlib.pyplot as plt
from sklearn import preprocessing
from sklearn.preprocessing import MinMaxScaler
```

```python
# Opening UAV data in may
UAV = xarray.open_dataarray("F:/Redaction/Bernadouze/Submitted/R2/Data_repository/BERNADOUZE_20230517_CUTFINAL_Mask.tif", engine="rasterio")
UAV.rio.to_raster("F:/Redaction/Bernadouze/Temp/temp.tif")
UAV = rasterio.open('F:/Redaction/Bernadouze/Temp/temp.tif')

# Reading bands
R = UAV.read(1)
G = UAV.read(2)
B = UAV.read(3)
RE = UAV.read(4)
NIR = UAV.read(5)

# Setting PAR and TEMP values
PAR = 0.33 # May
TEMP = 7.36 # May

# Calcuting indices
CVI = NIR * (R / (G * G))
NDRE = (NIR - RE) / (NIR + RE)
BWDRVI = ((0.1 * NIR) - B) / ((0.1 * NIR) + B)

# Calculating fluxes
NEE = 0.2 + (-3.7 * NDRE) + (-1.1 * BWDRVI) + (0.03 * TEMP)
ER = 1.6 + (-1.1 * BWDRVI) + (-4.1 * PAR) + (1.2 * NIR) + (0.05 * TEMP)
GEP = -2.0 + (-2.9 * NDRE) + (4.1 * PAR) + (0.6*PAR) + (-0.03 * TEMP)
UAV.close()

# Exporting
with rasterio.open(
    'F:/Redaction/Bernadouze/Submitted/R2/Data_repository/BERNADOUZE_20230517_CUTFINAL_Mask_NEE.tif',
    'w',
    driver='GTiff',
    height=UAV.shape[0],
    width=UAV.shape[1],
    count=1,
    dtype=NEE.dtype,
    crs= rasterio.crs.CRS.from_epsg(2154),
    transform=UAV.transform,
) as dst:
    dst.write(NEE, 1)
    
with rasterio.open(
    'F:/Redaction/Bernadouze/Submitted/R2/Data_repository/BERNADOUZE_20230517_CUTFINAL_Mask_ER.tif',
    'w',
    driver='GTiff',
    height=UAV.shape[0],
    width=UAV.shape[1],
    count=1,
    dtype=ER.dtype,
    crs= rasterio.crs.CRS.from_epsg(2154),
    transform=UAV.transform,
) as dst:
    dst.write(ER, 1)
    
with rasterio.open(
    'F:/Redaction/Bernadouze/Submitted/R2/Data_repository/BERNADOUZE_20230517_CUTFINAL_Mask_GEP.tif',
    'w',
    driver='GTiff',
    height=UAV.shape[0],
    width=UAV.shape[1],
    count=1,
    dtype=GEP.dtype,
    crs= rasterio.crs.CRS.from_epsg(2154),
    transform=UAV.transform,
) as dst:
    dst.write(GEP, 1)
```
