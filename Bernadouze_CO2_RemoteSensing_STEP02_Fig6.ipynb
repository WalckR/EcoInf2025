```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import csv
import matplotlib
import statsmodels.formula.api as smf
```

```python
# Load UAV dataset
dataset = pd.read_csv('F:/Redaction/Bernadouze/Figures/UAV_signature_May.csv',sep=';',index_col=0)
UAV_May = pd.DataFrame(dataset)

dataset = pd.read_csv('F:/Redaction/Bernadouze/Figures/UAV_signature_June.csv',sep=';',index_col=0)
UAV_June = pd.DataFrame(dataset)

dataset = pd.read_csv('F:/Redaction/Bernadouze/Figures/UAV_signature_July.csv',sep=';',index_col=0)
UAV_July = pd.DataFrame(dataset)

dataset = pd.read_csv('F:/Redaction/Bernadouze/Figures/UAV_signature_October.csv',sep=';',index_col=0)
UAV_October = pd.DataFrame(dataset)
```

```python
#plt.style.use('seaborn-v0_8-paper')
plt.style.use('default')
import matplotlib.transforms as mtransforms

fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(12, 12))

bplot1 = ax1.plot([450,560,650,730,840],UAV_May)
ax1.set_ylabel("Reflectance", fontweight='normal', fontsize=12)
ax1.set_ylim([0, 0.6])
ax1.legend(UAV_May.columns.values)
ax1.set_title('May',size=12)
trans = mtransforms.ScaledTranslation(-20/72, 7/72, fig.dpi_scale_trans)
ax1.text(-0.1, 1.0, 'a)', transform=ax1.transAxes + trans, fontsize = 14, va='bottom')

bplot2 = ax2.plot([450,560,650,730,840],UAV_June)
ax2.set_ylim([0, 0.6])
ax2.set_title('June',size=12)
trans = mtransforms.ScaledTranslation(-20/72, 7/72, fig.dpi_scale_trans)
ax2.text(-0.1, 1.0, 'b)', transform=ax2.transAxes + trans, fontsize = 14, va='bottom')

bplot3 = ax3.plot([450,560,650,730,840],UAV_July)
ax3.set_ylabel("Reflectance", fontweight='normal', fontsize=12)
ax3.set_ylim([0, 0.6])
ax3.set_title('July',size=12)
trans = mtransforms.ScaledTranslation(-20/72, 7/72, fig.dpi_scale_trans)
ax3.text(-0.1, 1.0, 'c)', transform=ax3.transAxes + trans, fontsize = 14, va='bottom')
ax3.set_xlabel("Wavelength (nm)", fontweight='normal', fontsize=12)

bplot4 = ax4.plot([450,560,650,730,840],UAV_October)
ax4.set_ylim([0, 0.6])
ax4.set_title('October',size=12)
trans = mtransforms.ScaledTranslation(-20/72, 7/72, fig.dpi_scale_trans)
ax4.text(-0.1, 1.0, 'd)', transform=ax4.transAxes + trans, fontsize = 14, va='bottom')
ax4.set_xlabel("Wavelength (nm)", fontweight='normal', fontsize=12)

plt.savefig("F:/Redaction/Bernadouze/Figures/Spectral_signatures.tif", dpi = 600)
```
